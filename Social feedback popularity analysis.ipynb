```python
from pyspark.sql import SparkSession

spark = SparkSession.\
        builder.\
        appName("HW2-2").\
        master("spark://spark-master:7077").\
        config("spark.executor.memory", "1g").\
        getOrCreate()
```

```python
from pyspark import SparkFiles
from pyspark.sql.types import *
import pandas as pd

Facebook_Economy = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/Facebook_Economy.csv")
Facebook_Microsoft = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/Facebook_Microsoft.csv")
Facebook_Obama = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/Facebook_Obama.csv")
Facebook_Palestine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/Facebook_Palestine.csv")
Facebook = pd.concat([Facebook_Economy, Facebook_Microsoft, Facebook_Obama, Facebook_Palestine])

display(Facebook)
```

```python
#Spark dataframe
pd.DataFrame.iteritems = pd.DataFrame.items
Facebook = spark.createDataFrame(Facebook)
Facebook.show()
```

```python
GooglePlus_Economy = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/GooglePlus_Economy.csv")
GooglePlus_Microsoft = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/GooglePlus_Microsoft.csv")
GooglePlus_Obama = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/GooglePlus_Obama.csv")
GooglePlus_Palestine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/GooglePlus_Palestine.csv")
GooglePlus = pd.concat([GooglePlus_Economy, GooglePlus_Microsoft, GooglePlus_Obama, GooglePlus_Palestine])

GooglePlus = spark.createDataFrame(GooglePlus)
GooglePlus.show()
```

```python
LinkedIn_Economy = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/LinkedIn_Economy.csv")
LinkedIn_Microsoft = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/LinkedIn_Microsoft.csv")
LinkedIn_Obama = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/LinkedIn_Obama.csv")
LinkedIn_Palestine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00432/Data/LinkedIn_Palestine.csv")
LinkedIn = pd.concat([LinkedIn_Economy, LinkedIn_Microsoft, LinkedIn_Obama, LinkedIn_Palestine])

LinkedIn = spark.createDataFrame(LinkedIn)
LinkedIn.show()
```

#### Average popularity of each news by hour, and by day, respectively for each platform.


```python
day1 = Facebook.columns[1:73]
day2 = Facebook.columns[73:145]
```

```python
import pyspark.sql.functions as func
from pyspark.sql.window import Window
f1 = Facebook.withColumn('day1', sum([func.col(c) for c in day1])/72).select('day1', 'IDLink')
f2 = Facebook.withColumn('day2', sum([func.col(c) for c in day1])/72).select('day2', 'IDLink')

#w = Window().orderBy(func.lit('A'))
f1.join(f2, on=['IDLink']).sort('IDLink', ascending=True).show()
```

```python
f1 = GooglePlus.withColumn('day1', sum([func.col(c) for c in day1])/72).select('day1', 'IDLink')
f2 = GooglePlus.withColumn('day2', sum([func.col(c) for c in day1])/72).select('day2', 'IDLink')

f1.join(f2, on=['IDLink']).sort('IDLink', ascending=True).show()
```

```python
f1 = LinkedIn.withColumn('day1', sum([func.col(c) for c in day1])/72).select('day1', 'IDLink')
f2 = LinkedIn.withColumn('day2', sum([func.col(c) for c in day1])/72).select('day2', 'IDLink')

f1.join(f2, on=['IDLink']).sort('IDLink', ascending=True).show()
```
